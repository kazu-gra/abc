#活性化関数

##シグモイド関数

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.arange(-5.0,5.0,0.1)
y=sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

##ソフトマックス関数

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
     exp_x = np.exp(x)
     sum_exp_x = np.sum(exp_x)
     return exp_x / sum_exp_x

x=np.arange(-5.0,5.0,0.1)
y=softmax(x)

plt.plot(x,y)
plt.ylim(-0.01,0.1)
plt.show()
```

#グリッド・凡例

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-np.pi,np.pi,0.1)

sin=np.sin(x)
cos=np.cos(x) 

plt.plot(x,sin, label='sin',color='blue',alpha=1.0)
plt.plot(x,cos, label='cos',color='red',alpha=0.2)

plt.grid() # グリッドの表示
plt.legend() # 凡例の表示
plt.show()
```

#課題１
tanh関数をグラフにしよう
 - 条件： x = np.arange(-5.0, 5.0, 0.1)， plt.ylim(-1.2,1.2)，タイトル，軸ラベル

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

x=np.arange(-5.0,5.0,0.1)
y=tanh(x)

plt.plot(x,y)
plt.ylim(-1.2,1.2)
plt.title('tanh')#タイトル
plt.xlabel('input')#x軸ラベル
plt.ylabel('output')#y軸ラベル
plt.show()
```
